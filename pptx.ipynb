```python
import os
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN

# Create a new presentation
prs = Presentation()

########################################################################################################### SLIDE 1: Title Slide
title_slide_layout = prs.slide_layouts[0]
slide1 = prs.slides.add_slide(title_slide_layout)
slide_width = prs.slide_width.inches
title = slide1.shapes.title
subtitle = slide1.placeholders[1]

title.text = "QC CHECK -TEXT CLASH"
# Move title down by 1.5 inches
title.top = title.top + Inches(1.5)
# Preserve the original title width for centering, then force it to 9 inches (if desired)
title_width = title.width.inches
title.width = Inches(9)
title.left = Inches((slide_width - title_width) / 2)
subtitle.text = "2025"

text = "SMART TECHNOLOGY FOR SMARTER MOBILITY"
font_size = Pt(20)
char_width_in_inches = 0.5 * font_size.pt / 72
text_width = char_width_in_inches * len(text)  # Estimated width of the entire text in inches
text_height = 1  # Fixed height (in inches) for the textbox

# Center the textbox on the slide
left_box = Inches((slide_width - text_width) / 2)
top_box = Inches(7)
width_box = Inches(text_width)
height_box = Inches(text_height)
txBox = slide1.shapes.add_textbox(left_box, top_box, width_box, height_box)
tf = txBox.text_frame
tf.clear()
p = tf.paragraphs[0]
p.text = text
p.alignment = PP_ALIGN.CENTER

############################################################################################################# SLIDE 2: Blank Slide 
blank_slide_layout = prs.slide_layouts[6]
slide2 = prs.slides.add_slide(blank_slide_layout)

text = "OVERLAPPED TEXTS IN SHEET - "
font_size = Pt(20)
char_width_in_inches = 0.5 * font_size.pt / 72
text_width = char_width_in_inches * len(text)
text_height = 1
left_box = Inches((slide_width - text_width) / 2)
top_box = Inches(0.25)
width_box = Inches(text_width)
height_box = Inches(text_height)
txBox = slide2.shapes.add_textbox(left_box, top_box, width_box, height_box)
tf = txBox.text_frame
tf.clear()
p = tf.paragraphs[0]
p.text = text
p.alignment = PP_ALIGN.CENTER

# Add a picture to slide 2
img_path = r"c:\Users\jjers\OneDrive\Pictures\Paint 3D\Night Sky.PNG"   # Ensure this path is correct.
img_width = Inches(9)
img_height = Inches(4.5)
left_img = Inches((slide_width - 9) / 2)
top_img = Inches(1)
slide2.shapes.add_picture(img_path, left_img, top_img, width=img_width, height=img_height)

########################################################################################################################## SLIDE 3:
slide3 = prs.slides.add_slide(blank_slide_layout)
slide3_width = prs.slide_width.inches
slide3_height = prs.slide_height.inches

img_paths_slide3 = [
    r"c:\Users\jjers\OneDrive\Pictures\Image1.PNG",
    r"c:\Users\jjers\OneDrive\Pictures\Image2.PNG",
    r"c:\Users\jjers\OneDrive\Pictures\Image3.PNG",
    r"c:\Users\jjers\OneDrive\Pictures\Image4.PNG",
    r"c:\Users\jjers\OneDrive\Pictures\Image5.PNG",
    r"c:\Users\jjers\OneDrive\Pictures\Image6.PNG",
]

gap = 0.1  # Gap in inches
cols = 3
rows = 2

cell_width = (slide3_width - (cols + 1) * gap) / cols
cell_height = (slide3_height - (rows + 1) * gap) / rows
for index, img in enumerate(img_paths_slide3):
    row = index // cols   # Integer division gives row number (0 or 1)
    col = index % cols    # Modulo gives column index (0, 1, or 2)
    left_img = gap + col * (cell_width + gap)
    top_img = gap + row * (cell_height + gap)
    
    slide3.shapes.add_picture(
        img,
        Inches(left_img),
        Inches(top_img),
        width=Inches(cell_width),
        height=Inches(cell_height)
    )

############################################################################################################################

file_name = 'txt_clash.pptx'
prs.save(file_name)
os.startfile(file_name)
```
